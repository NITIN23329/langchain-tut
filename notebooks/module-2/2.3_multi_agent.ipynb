```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain.tools import tool


@tool
def add(a:int, b: int)-> int:
    '''return addition of two numbers'''
    return a + b

@tool
def subtract(a:int, b: int)-> int:
    '''return subtraction of two numbers'''
    return a - b

```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
model = ChatGoogleGenerativeAI(model = 'gemini-3.1-flash-lite-preview')
```

```python
from langchain.agents import create_agent

sub_agent1 = create_agent(model = model, tools = [add])
sub_agent2 = create_agent(model = model, tools = [subtract])
```

```python
from langchain.messages import HumanMessage
@tool
def invoke_sub_agent1(a: int, b: int)-> int: 
    '''call subagent1 to calculate the addition of two numbers'''
    response = sub_agent1.invoke({'messages' : [HumanMessage(f'Calculate addition of {a} and {b}')]})
    return response['messages'][-1].content
@tool
def invoke_sub_agent2(a: int, b: int)-> int: 
    '''calcualte subagent2 to calculate the subtraction of two numbers'''
    response = sub_agent2.invoke({'messages' : [HumanMessage(f'Calculate subtraction of {a} and {b}')]})
    return response['messages'][-1].content    
```

```python
agent = create_agent(model = model, tools = [invoke_sub_agent1, invoke_sub_agent2])
```

```python
response = agent.invoke({'messages' : [HumanMessage('What is 5 plus 6')]})
```
