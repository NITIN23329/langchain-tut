```python
from langchain_google_genai import ChatGoogleGenerativeAI
model = ChatGoogleGenerativeAI(model = 'gemini-3.1-flash-lite-preview')
```

```python
from langchain.agents import AgentState
from dataclasses import dataclass

@dataclass
class UserContext(AgentState):
    lang: str = 'English' #default state value
```

```python
from langchain.agents.middleware import ModelRequest, ModelResponse, wrap_model_call
from collections.abc import Callable
from langchain.messages import SystemMessage
from pprint import pprint
@wrap_model_call
def get_dynamic_prompt(model_request: ModelRequest, handler: Callable[[ModelRequest], ModelResponse])-> ModelResponse:
    '''generate sysem prompt based on user language'''
    sys_prompt = 'You are an helpful assistant'
    pprint(f'model request: {model_request}')
    lang = model_request.state['lang']
    if lang == 'English':
        return handler(model_request.override(system_message=sys_prompt))
    return  handler(model_request.override(system_message=sys_prompt + f'. Please reponse in {lang} language to user')) 
```

```python
from langchain.agents import create_agent
agent = create_agent(model = model, middleware = [get_dynamic_prompt], state_schema = UserContext)
```

```python
from langchain.messages import HumanMessage

response = agent.invoke({'messages':[HumanMessage('Hi how are you?')], 'lang': 'Hindi'})
```

```python
###dynamic model 
```

```python
model_small = ChatGoogleGenerativeAI(model = 'gemini-2.5-flash-lite')
model_large = ChatGoogleGenerativeAI(model = 'gemini-3.1-flash-lite-preview')
```

```python
from langchain.agents import AgentState
from dataclasses import dataclass
@dataclass
class UserContext(AgentState):
    user_type: str # free, paid
```

```python
from langchain.agents.middleware import wrap_model_call, ModelRequest, ModelResponse
from collections.abc import Callable
from pprint import pprint
@wrap_model_call
def dynamic_model(request : ModelRequest, handler: Callable[[ModelRequest], ModelResponse])-> ModelResponse: 
    '''function to set tools for agent dynamically'''
    pprint(request)
    user_type = request.state['user_type']
    model = model_small
    if user_type == 'paid':
        model = model_large
    return handler(request.override(model=model))
```

```python
from langchain.agents import create_agent
agent = create_agent(model=model_small, middleware=[dynamic_model], state_schema = UserContext)
```

```python
from langchain.messages import HumanMessage

response = agent.invoke({'messages':[HumanMessage('Hi how are you?')], 'user_type': 'paid'})
```

```python
response
```

```python
from langchain.tools import tool

@tool
def normal_user()->str: 
    '''return a good greeting string for user and offer infomration'''
    return 'Hi Your offer is 10% on all product'
@tool
def paid_user_dynamic()->str: 
    '''return a good greeting string for paid user'''
    return 'Hi Paid user, thanks for being a valuable customer. Your offer is 80% on all product'
@tool 
def free_user_dynamic()-> str:
    '''asking free user to upgrate and greeting'''
    return 'Hi Free user, Your offer is 20% on all product'
```

```python
from langchain.agents.middleware import wrap_model_call, ModelRequest, ModelResponse
from collections.abc import Callable
from pprint import pprint
@wrap_model_call
def dynamic_tool(request : ModelRequest, handler: Callable[[ModelRequest], ModelResponse])-> ModelResponse: 
    '''function to set tools for agent dynamically'''
    pprint(request)
    user_type = request.state['user_type']
    tools = []
    if user_type == 'paid':
        tools = [paid_user_dynamic,free_user_dynamic]
    else: 
        tools = [normal_user]
    return handler(request.override(tools=tools))
```

```python
from langchain.agents import create_agent
agent = create_agent(model=model, middleware=[dynamic_tool],tools=[normal_user,paid_user_dynamic,free_user_dynamic], state_schema = UserContext, system_prompt='use tool call to get offer')
```

```python
from langchain.messages import HumanMessage

response = agent.invoke({'messages':[HumanMessage('Hi what offer avaiable for me? I am paid user')], 'user_type': 'paid'})
```
