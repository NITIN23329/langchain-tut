```python
from dotenv import load_dotenv

load_dotenv()
```

```python
### calling chat model
```

```python
from langchain.chat_models import init_chat_model
```

```python
model = init_chat_model( model = 'gemini-3.1-flash-lite-preview', model_provider = 'google_genai', temperature =  0.8) # need to add model_provider = 'google_genai' otherwise google try to use google-vertexAI instead of our google-genai
```

```python
response = model.invoke("What is the capital of India")
response
```

```python
response.content
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI

model = ChatGoogleGenerativeAI(model="gemini-3.1-flash-lite-preview")

response = model.invoke("What's the capital of the Moon?")
print(response)
```

```python
### creating a simple agent
```

```python
from langchain.agents import create_agent
```

```python
agent = create_agent(model = model)
```

```python
from langchain.messages import HumanMessage, SystemMessage, AIMessage

# this the format of langchain agents instead of role and content
messages = [
    SystemMessage(content='You are a science fiction writer, you can create fictional objects in the space'),
    HumanMessage(content='What is the capital of Moon'),
    AIMessage(content='The capital of moon is Luna'),
    HumanMessage(content='Can you tell me more about luna'),
]
```

```python
response = agent.invoke({'messages': messages})
```

```python
response['messages'][-1].content
```
