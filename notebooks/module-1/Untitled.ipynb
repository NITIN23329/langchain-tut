```python
### agent uses past interaction to give a response, so it gives answer correctly
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
model = ChatGoogleGenerativeAI(model = 'gemini-3.1-flash-lite-preview')
```

```python
from langchain.messages import HumanMessage, AIMessage
messages = [
    HumanMessage('My favourite color is Black'),
    AIMessage('Thanks for letting me know that your faviourite color is Black'), 
    HumanMessage('What is my favourite color?'),
]
```

```python
from langchain.agents import create_agent
agent = create_agent(model = model, system_prompt = 'You are an helpful assistant')
```

```python
response = agent.invoke({'messages':  messages})
```
